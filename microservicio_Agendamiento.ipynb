### **Se realizan las peticiones al servidor de la UIS con los codigos de las materias, esto con la intension de acpturar los nombres de estas y asi poder hacer la ultima peticion que nos peritira saber los horarios de las materias, el lugar y el porfesor que la dicta**

```python
def peticionAsignaturaProgrmadaUIS(codigo):
  URL = f"https://www.uis.edu.co/estudiantes/asignaturas_programadas/resultado_buscador.jsp?nombre=&codigo={codigo}&parametro=2 "
  page1 = requests.get(URL)
  soup1 = BeautifulSoup(page1.content, "html.parser")

  data =[]
  data_row = []
  for item in  soup1.find_all('td'):
    row = item.text.replace('\n','').strip()
    row = row.replace('\r','')
    row = row.replace('\t','') 
    if ("HORARIOS DISPONIBLES PARA LA ASIGNATURA"in row):
      continue
    if (row == ""):
      continue   
    if ("Matriculados" in row):
      data_row.append(row)
      data.append(data_row)
      data_row = []
      continue
    data_row.append(row)

  definitive_list =[]
  for fila in data:
    number = 0
    provisional_dict = {}
    for element in fila:
      if (number<=1):
        if(("\xa0" not in element) and number == 0 and (":" not in element)):
          lista_matricula = []
          lista_matricula.append('Materia')
          lista_matricula.append(element)
          provisional_dict[lista_matricula[0]]=lista_matricula[1]
      if (number>=1):
        if ("\xa0" in element):
          list_diction = element.split("\xa0")
          provisional_dict[list_diction[0]]=list_diction[1]   
        else:    
          list_diction = element.split(": ")
          provisional_dict[list_diction[0]]=list_diction[1]
      number += 1
    definitive_list.append(provisional_dict)
    agendamiento_df = pd.DataFrame(definitive_list)

  return agendamiento_df
```

```python
def encontrarNombre(asinaturasProgramadas_df):
  cols_nombreMateria_df = ['Materia','Código']
  nombreMateria_df = asinaturasProgramadas_df[cols_nombreMateria_df]
  nombreMateria_df = nombreMateria_df.drop_duplicates(subset = ["Materia"])
  nombreMateria_df = nombreMateria_df['Materia']
  nombreMateria = nombreMateria_df.to_string()
  nombreMateria = nombreMateria.replace("0","").strip()
  return nombreMateria
```

#Peticion a pagina de la UIS -> Estudiantes -> asignaturas programadas -> horaria asignatura. Dando los parametros de **codigo, grupo, nombre** de la asignatura

```python
def BucarHorarioMateria(codigo, grupo, nombre):

  URL = f"https://www.uis.edu.co/estudiantes/asignaturas_programadas/resultado_buscador.jsp?nombre=&codigo={codigo}&parametro=2"
  page1 = requests.get(URL)
  soup1 = BeautifulSoup(page1.content, "html.parser")

  data =[]
  data_row = []
  for item in  soup1.find_all('td'):
    row = item.text.replace('\n','').strip()
    row = row.replace('\r','')
    row = row.replace('\t','') 
    if ("HORARIOS DISPONIBLES PARA LA ASIGNATURA"in row):
      continue
    if (row == ""):
      continue   
    if ("Matriculados" in row):
      data_row.append(row)
      data.append(data_row)
      data_row = []
      continue
    data_row.append(row)


  URL2 = f"https://www.uis.edu.co/estudiantes/asignaturas_programadas/horario_asignatura.jsp?codigo={codigo}&grupo={grupo}&nombre={nombre}"
  page = requests.get(URL2)
  soup = BeautifulSoup(page.content, "html.parser")
  data =[]
  data_row = []
  dias = ("LUNES","MARTES","MIERCOLES","JUEVES","VIERNES","SABADO","DOMINGO")
  print("---------------------------------------------------------------------------")
  Calendario = []
  contador= 0
  cabecera = ""
  apartir_de= True
  for item in soup.find_all("td"):
    if (apartir_de):
      cabecera = cabecera + " " +str(row)
      if (row == "Día"):
        apartir_de = False
    row = item.text.replace('\n','').strip()
    row = row.replace('\r','')
    row = row.replace('\t','') 
    if (" - "in row):
      row = row.replace(" ", "")
      row = row.replace("-", " - ")
    if (row == "Horario" or row == "Profesor" or row == "Edificio"):
      continue
    if (row in dias or (contador != 0 and contador<=3)):
      Calendario.append(row)
      contador += 1
    if (row == "Día" or item == soup.find_all("td")[-1]):
      contador = 0
      if ( len(Calendario) != 0):
        data.append(Calendario)
        Calendario = []  
  print(cabecera)
  print("---------------------------------------------------------------------------")
  result_df = pd.DataFrame(data, columns= ["DiaSemana", "Horario", "Lugar", "Profesor"])
  result_df[['Edificio','Salon']] = result_df.Lugar.str.split("-",expand=True)
  result_df = result_df.drop(columns=['Lugar'])
  result_df['IdMat'] = codigo
  result_df['Grupo'] = grupo
  result_df['nombre'] = nombre
  return result_df
```

```python
def insercionDB(dataframe, nombreTabla, cursorConec):
  cols = "`,`".join([str(i) for i in dataframe.columns.tolist()])
  # Insertar los registros del DataFrame uno a uno.
  for i,row in dataframe.iterrows():
      sql = f"INSERT INTO `{nombreTabla}` (`" +cols + "`) VALUES (" + "%s,"*(len(row)-1) + "%s)"
      cursorConec.execute(sql, tuple(row))
      # commit para guardar cambios
      connection.commit()
```

```python
def selectALLDB(nombreTabla, cursorConec):  
  # Revision de que los registros quedaran guardados
  sql = f"SELECT * FROM `{nombreTabla}`"
  cursorConec.execute(sql)

  # Fetch all los registros de la tabla e impresion uno a uno para verlos
  result = cursorConec.fetchall()
  for i in result:
      print(i)
```
